```python
df = df.filter(['Text','Label'], axis=1)
df['prefix'] = 'binary classification'
df['input_text'] = df['Text']
df['target_text'] = df['Label']
df = df.drop(['Label'], axis = 1)
```

```python
df = df.drop(['Text'], axis = 1)
df['target_text'] = df['target_text'].astype(str)
```

```python
df
```

```python
train_df = df[:int(0.8*len(df))]
test_df = df[int(0.8*len(df)):]
```

```python
 model_args = {
        "reprocess_input_data": True,
        "overwrite_output_dir": True,
        "max_seq_length": 10,
        "train_batch_size": 8,
        "num_train_epochs": 5,
        "save_model_every_epoch": False,
        "max_length": 512,
        "num_beams": 1,
    }


model = T5Model("t5", "t5-base", args=model_args, use_cuda=False)
```

```python
model.train_model(train_df)
```

```python
def count_matches(labels, preds):
    print(labels)
    print(preds)
    return sum([1 if label == pred else 0 for label, pred in zip(labels, preds)])

print(model.eval_model(test_df, matches=count_matches))
```
